```python
# File location and type
file_location = "/FileStore/tables/Balaji_Fast_Food_Sales.csv"
file_type = "csv"

# CSV options
infer_schema = "false"
first_row_is_header = "false"  # Change this to 'true' if the first row is the header
delimiter = ","

# Load the CSV into a Spark DataFrame with correct options
df = spark.read.format(file_type) \
    .option("inferSchema", infer_schema) \
    .option("header", first_row_is_header) \
    .option("sep", delimiter) \
    .load(file_location)

# Renaming the default columns (_c0, _c1, ...) to meaningful column names
df = df.withColumnRenamed('_c0', 'order_id') \
       .withColumnRenamed('_c1', 'date') \
       .withColumnRenamed('_c2', 'item_name') \
       .withColumnRenamed('_c3', 'item_type') \
       .withColumnRenamed('_c4', 'item_price') \
       .withColumnRenamed('_c5', 'quantity') \
       .withColumnRenamed('_c6', 'transaction_amount') \
       .withColumnRenamed('_c7', 'transaction_type') \
       .withColumnRenamed('_c8', 'received_by') \
       .withColumnRenamed('_c9', 'time_of_sale')

# Display the updated DataFrame with proper column names
df.show(5)

# Display the DataFrame columns to verify the correct structure
print(df.columns)

# If you want to use SQL queries, you can create a temporary view
df.createOrReplaceTempView("FastFoodSales")

# Example SQL query on the DataFrame
result = spark.sql("SELECT * FROM FastFoodSales WHERE transaction_type IS NOT NULL")
result.show(5)
```

```python
from pyspark.sql.functions import month, year, quarter, to_date, sum as sum_, count, desc, col, dense_rank, row_number
from pyspark.sql.types import IntegerType
from pyspark.sql.window import Window

# Convert date column to date type
df = df.withColumn("date", to_date(df.date, "MM/dd/yyyy"))

# Add year, month, and quarter columns
df = df.withColumn("year", year(df.date))
df = df.withColumn("month", month(df.date))
df = df.withColumn("quarter", quarter(df.date))

# Convert item_price and transaction_amount to integer type
df = df.withColumn("item_price", df.item_price.cast(IntegerType()))
df = df.withColumn("transaction_amount", df.transaction_amount.cast(IntegerType()))

# 1. Most Ordered Product
most_ordered = df.groupBy("item_name").agg(sum_("quantity").alias("total_quantity")).orderBy(desc("total_quantity")).limit(1)
display(most_ordered)

# 2. Total Amount Spent by each customer
customer_spend = df.groupBy("order_id").agg(sum_("transaction_amount").alias("total_spent")).orderBy("order_id")
display(customer_spend)

# 3. Total Amount spent by each food category
category_spend = df.groupBy("item_type").agg(sum_("transaction_amount").alias("total_spent")).orderBy(desc("total_spent"))
display(category_spend)

# 4. Yearly Sales
yearly_sales = df.groupBy("year").agg(sum_("transaction_amount").alias("total_sales")).orderBy("year")
display(yearly_sales)

# 5. Monthly Sales
monthly_sales = df.groupBy("year", "month").agg(sum_("transaction_amount").alias("total_sales")).orderBy("year", "month")
display(monthly_sales)

# 6. Quarterly Sales
quarterly_sales = df.groupBy("year", "quarter").agg(sum_("transaction_amount").alias("total_sales")).orderBy("year", "quarter")
display(quarterly_sales)

# 7. Frequency of Customer Visits
# Assuming each order_id is a unique visit
customer_frequency = df.groupBy("order_id").agg(count("*").alias("visit_count")).groupBy("visit_count").agg(count("*").alias("frequency")).orderBy("visit_count")
display(customer_frequency)

# 8. Total Sales by Source (assuming transaction_type is the source)
sales_by_source = df.groupBy("transaction_type").agg(sum_("transaction_amount").alias("total_sales")).orderBy(desc("total_sales"))
display(sales_by_source)

# 9. Top Ordered Items
top_items = df.groupBy("item_name").agg(sum_("quantity").alias("total_quantity"), sum_("transaction_amount").alias("total_sales"))
windowSpec = Window.orderBy(desc("total_quantity"))
top_items = top_items.withColumn("rank", dense_rank().over(windowSpec)).filter(col("rank") <= 5).orderBy(desc("total_quantity"))
display(top_items)

# 10. Total Sales by each country (assuming we don't have country data, we'll use received_by as a proxy)
sales_by_country = df.groupBy("received_by").agg(sum_("transaction_amount").alias("total_sales")).orderBy(desc("total_sales"))
display(sales_by_country)

# Additional Analysis: Sales Trend
sales_trend = df.groupBy("date").agg(sum_("transaction_amount").alias("daily_sales")).orderBy("date")
display(sales_trend)

# Additional Analysis: Average Transaction Amount
avg_transaction = df.groupBy("order_id").agg(sum_("transaction_amount").alias("transaction_total"))
avg_transaction = avg_transaction.agg({"transaction_total": "avg"}).withColumnRenamed("avg(transaction_total)", "average_transaction_amount")
display(avg_transaction)

# Additional Analysis: Best Selling Items by Revenue
best_selling_revenue = df.groupBy("item_name").agg(sum_("transaction_amount").alias("total_revenue")).orderBy(desc("total_revenue")).limit(5)
display(best_selling_revenue)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pyspark.sql import SparkSession
from pyspark.sql.functions import year, month, quarter, sum as spark_sum

# Initialize Spark session
spark = SparkSession.builder.appName("SalesAnalysis").getOrCreate()

# Handle date formatting if needed (adjust the format to match your dataset)
from pyspark.sql.functions import to_date
df_spark = df_spark.withColumn("date", to_date(df_spark['date'], "MM/dd/yyyy"))

# Extract year, month, and quarter from the date
df_spark = df_spark.withColumn("year", year(df_spark['date']))
df_spark = df_spark.withColumn("month", month(df_spark['date']))
df_spark = df_spark.withColumn("quarter", quarter(df_spark['date']))

# Data aggregations for different plots

# 1. Most Ordered Product
most_ordered_product = df_spark.groupBy('item_name').agg(spark_sum('quantity').alias('total_quantity')).orderBy('total_quantity', ascending=False).limit(1)
most_ordered_pd = most_ordered_product.toPandas()

# 2. Total Amount Spent by Each Customer
total_spent_customer = df_spark.groupBy('order_id').agg(spark_sum('transaction_amount').alias('total_spent'))
total_spent_customer_pd = total_spent_customer.toPandas()

# 3. Total Amount Spent by Each Food Category
category_spent = df_spark.groupBy('item_type').agg(spark_sum('transaction_amount').alias('total_spent')).orderBy('total_spent', ascending=False)
category_spent_pd = category_spent.toPandas()

# 4. Yearly Sales
yearly_sales = df_spark.groupBy('year').agg(spark_sum('transaction_amount').alias('total_sales'))
yearly_sales_pd = yearly_sales.toPandas()

# 5. Monthly Sales
monthly_sales = df_spark.groupBy('month').agg(spark_sum('transaction_amount').alias('total_sales'))
monthly_sales_pd = monthly_sales.toPandas()

# 6. Quarterly Sales
quarterly_sales = df_spark.groupBy('quarter').agg(spark_sum('transaction_amount').alias('total_sales'))
quarterly_sales_pd = quarterly_sales.toPandas()

# 7. Frequency of Customer Visits (Assuming each order_id is a customer visit)
customer_visits = df_spark.groupBy('order_id').count()
customer_visits_pd = customer_visits.toPandas()

# 8. Total Sales by Source (transaction_type: Cash/Online)
sales_by_source = df_spark.groupBy('transaction_type').agg(spark_sum('transaction_amount').alias('total_sales'))
sales_by_source_pd = sales_by_source.toPandas()

# 9. Top Ordered Items
top_ordered_items = df_spark.groupBy('item_name').agg(spark_sum('quantity').alias('total_quantity')).orderBy('total_quantity', ascending=False).limit(5)
top_ordered_items_pd = top_ordered_items.toPandas()

# Plot the dashboard using matplotlib
fig, axs = plt.subplots(3, 3, figsize=(15, 12))
fig.suptitle('Sales Analysis Dashboard', fontsize=16)

# 1. Most Ordered Product
axs[0, 0].barh(most_ordered_pd['item_name'], most_ordered_pd['total_quantity'], color='lightgreen')
axs[0, 0].set_title('Most Ordered Product')

# 2. Total Amount Spent by Each Customer
axs[0, 1].scatter(total_spent_customer_pd['order_id'], total_spent_customer_pd['total_spent'], color='lightblue')
axs[0, 1].set_title('Total Amount Spent by Each Customer')

# 3. Total Amount Spent by Each Food Category
axs[0, 2].bar(category_spent_pd['item_type'], category_spent_pd['total_spent'], color='lightgreen')
axs[0, 2].set_title('Total Amount Spent by Each Food Category')

# 4. Yearly Sales
axs[1, 0].plot(yearly_sales_pd['year'], yearly_sales_pd['total_sales'], marker='o', color='purple')
axs[1, 0].set_title('Yearly Sales')

# 5. Monthly Sales
axs[1, 1].bar(monthly_sales_pd['month'], monthly_sales_pd['total_sales'], color='lightgreen')
axs[1, 1].set_title('Monthly Sales')

# 6. Quarterly Sales (Pie Chart)
axs[1, 2].pie(quarterly_sales_pd['total_sales'], labels=quarterly_sales_pd['quarter'], autopct='%1.1f%%', colors=sns.color_palette('muted'))
axs[1, 2].set_title('Quarterly Sales')

# 7. Frequency of Customer Visits
axs[2, 0].hist(customer_visits_pd['count'], bins=10, color='purple')
axs[2, 0].set_title('Frequency of Customer Visits')

# 8. Total Sales by Source
axs[2, 1].pie(sales_by_source_pd['total_sales'], labels=sales_by_source_pd['transaction_type'], autopct='%1.1f%%', colors=['gold', 'skyblue'])
axs[2, 1].set_title('Total Sales by Source')

# 9. Top Ordered Items
axs[2, 2].bar(top_ordered_items_pd['item_name'], top_ordered_items_pd['total_quantity'], color='lightgreen')
axs[2, 2].set_title('Top Ordered Items')

plt.tight_layout()
plt.subplots_adjust(top=0.9)
plt.show()
```
